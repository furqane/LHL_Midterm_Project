```python
import pandas as pd
```

```python
df = pd.read_csv('ftest_to_encode.csv', index_col=[0])
df = df.drop(['mkt_unique_carrier'],axis=1) #this feature was not used in XGBoost
df.reset_index(drop=True, inplace=True)
df
```

```python
predicted_df = pd.read_csv('predicted_df.csv', index_col=[0])
predicted_df.reset_index(drop=True, inplace=True)
predicted_df
```

```python
submission_df = pd.concat([df,predicted_df],axis=1)
```
